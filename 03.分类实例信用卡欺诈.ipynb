```python
import torch
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./dataset/credit-a.csv",header=None)
data
```

```python
# 数据项，切掉最后一列的标签项
X = data.iloc[:,:-1]
```

```python
X
```

```python
Y = data.iloc[:,-1].replace(-1,0)
Y
```

```python
Y.value_counts()
```

```python
X = torch.from_numpy(X.values).type(torch.FloatTensor)
X.shape
```

```python
Y = torch.from_numpy(Y.values.reshape(-1,1)).type(torch.FloatTensor)
```

```python
Y.size()
```

```python
from torch import nn
```

```python
model = nn.Sequential(nn.Linear(15,1),nn.Sigmoid())
```

```python
model
```

```python
loss_fn = nn.BCELoss()
```

```python
opt = torch.optim.SGD(model.parameters(),lr=0.00001)
```

```python
batches = 16
no_of_batches = 654//16
```

```python
for epoch in range(1000):
    for batch in range(no_of_batches):
        start = batch * batches
        end = start + batch
        x = X[start:end]
        y = Y[start:end]
        y_pred = model(x)
        loss = loss_fn(y_pred,y)
        opt.zero_grad()
        loss.backward()
        opt.step()
```

```python
model.state_dict()
```

## 计算正确率

```python
((model(X).data.numpy() > 0.5).astype('int') == Y.numpy()).mean()
```
