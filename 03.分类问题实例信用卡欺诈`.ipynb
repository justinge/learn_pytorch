```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("./dataset/credit-a.csv")
data.head(5)
```

```python
data = pd.read_csv("./dataset/credit-a.csv",header=None)
data.head(5)
```

```python
# 拆分数据X和y
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
# 把-1 替换成0
Y = data.iloc[:,-1].replace(-1, 0)
Y
```

```python
Y.value_counts()
```

```python
X = torch.from_numpy(X.values).type(torch.FloatTensor)
```

```python
X=X.type(torch.FloatTensor)
X
```

```python
Y = torch.from_numpy(Y.values.reshape(-1, 1)).type(torch.FloatTensor)
Y
```

```python
# 因为是二分类问题，所以最终需要将线性计算结果，拟合到0，1之间，用sigmoid函数，因为15个特征，所以选择15，输出的是0，1之间的概率，就是1个特征
```

```python
model = nn.Sequential(
    nn.Linear(15,1),
    nn.Sigmoid()
    )
```

```python
model
```

```python
# 二元交叉熵, 二分类 
loss_fn = nn.BCELoss()
```

```python
# 目标函数,设置目标函数的优化策略
opt = torch.optim.Adam(model.parameters(), lr=0.0001)
opt
```

```python
batches = 16
# 整除的方式, 不足一个batch的就舍掉,比如0.25
no_of_batches = len(Y)//16
```

```python
for epoch in range(1000):
    total_loss = 0
    for batch in range(no_of_batches):
        # 起始位置
        start = batch*batches
        end = start + batches
        # start = batch*batches
        # end = start + batches
        x = X[start: end]
        y = Y[start: end]
        # 计算loss
        y_pred = model(x)
        loss = loss_fn(y_pred, y)
        total_loss+=loss.item()
        opt.zero_grad()
        loss.backward()
        opt.step()
        # 实际预测结果
    print(epoch,"acc",((model(X).data.numpy() > 0.5).astype('int')==Y.numpy()).mean())
    print(epoch,"loss",total_loss)
```
